```python
import pandas as pd
```

```python
ilks_df = pd.read_csv("./data/ilk_data.csv")
ilks_df
```

```python
dog_df = pd.read_csv("./data/dog_events.csv")
dog_df["symbol"] = dog_df.event_ilk.apply(lambda x: ilks_df[ilks_df["ilk"]==x].symbol.iloc[0])
dog_df["name"] = dog_df.event_ilk.apply(lambda x: ilks_df[ilks_df["ilk"]==x].name.iloc[0])
dog_df["datetime"] = pd.to_datetime(dog_df['timestamp'].apply(lambda x: (x+18030)*(10**9))) # Datetime in IST
dog_df
```

```python
dog_df["from"].mode() # The address which called bark in dog contract the most
```

```python
dog_df.groupby('from').count().sort_values("symbol",ascending=False)["symbol"] # Address in the frequency order of the number of times they have successfully called bark on the dog contract
```

```python
from_address = dog_df.groupby('from').count().sort_values("symbol",ascending=False).index[2] #Bark events of a certain address in the frequency ranking order
dog_df[dog_df["from"]==from_address]
```

```python
dog_df[dog_df["from"]==from_address].iloc[-1]["blockNumber"]
```

```python
clip_df = pd.read_csv("./data/clip_events.csv")
clip_df["datetime"] = pd.to_datetime(clip_df['timestamp'].apply(lambda x: (x+18030)*(10**9))) # Datetime in IST
clip_df
```

```python
def compare_froms(x):
    cf = clip_df[clip_df.blockNumber==x["blockNumber"]][clip_df.transactionIndex==x["transactionIndex"]]["from"]
    if len(cf) > 0:
        if cf.iloc[0] == x["from"]:
            return 1
        else:
            return 0
    else:
        return -1

dog_df["compare_froms"] = dog_df.apply(lambda x: compare_froms(x), axis=1)
len(dog_df[dog_df["compare_froms"]>=0]),len(dog_df[dog_df["compare_froms"]==0])
```
