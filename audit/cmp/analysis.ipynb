```python
models = {}
for k,v in data['InterestRateModel'].items():
    mod = v["contract"]
    if mod in models:
        models[mod].append((k,v["address"]))
    else:
        models[mod] = [(k,v["address"])]
for mod, modval in models.items():
    print("Model: ", mod, "\n")
    for val in modval:
        print(val[0], val[1])
    print("\n")
```

```python
import pandas as pd
```

```python
df = pd.read_csv("G:/crypto/smart/defi/cmp/tokenIrms.csv")
df
```

```python
irms = set(df.irm_name.values)
irm_tokens = {}
for irm in irms:
    irm_tokens[irm] = list(df[df.irm_name==irm].token.values)
irm_tokens
```

```python
from os import listdir as ls
```

```python
reading_folders = ["WhitePaperInterestRateModel", "DAIInterestRateModel", "DAIInterestRateModelV2", "DAIInterestRateModelV3",
                 "JumpRateModel", "JumpRateModelV2", "LegacyJumpRateModelV2"]
for fold in reading_folders:
    fils = [dei.split(".sol")[0] for dei  in ls(f"./contracts/InterestModels/{fold}")]
    order = []
    for fil in fils:
        if fil in irms:
            order.append(fil)
    if len(order) > 0:
        print(fold, "\n")
        for fil in order:
            print(fil, *irm_tokens[fil])
        print("\n")
```
